```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ARA(2019-2023) - Principais causas de internação - Principais causas de internação.csv')
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
doencas = df.iloc[:14]
```

```python
doencas
```

```python
df_doencas = doencas.melt(id_vars='Causas', var_name='Ano', value_name='Casos')
```

```python
df_doencas
```

```python
# Salvar o DataFrame como um arquivo CSV
df_doencas.to_csv('dados_doencas.csv', index=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_doencas, x='Ano', y='Casos', hue='Causas', palette='Set1')
plt.title('Evolução dos Casos de Doenças ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Casos')
plt.legend(title='Doença', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
