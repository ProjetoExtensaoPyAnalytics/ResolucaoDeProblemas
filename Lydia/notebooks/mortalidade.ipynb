```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ARA(2019-2022) - Mortalidade por grupos de causas -  Mortalidade por grupos de causas.csv')
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
obitos = df[:14]
```

```python
obitos
```

```python
df_obitos = obitos.melt(id_vars='Causas', var_name='Ano', value_name='Óbitos')
```

```python
df_obitos
```

```python
df_obitos['Óbitos'] = df_obitos['Óbitos'].astype(int)
```

```python
df_obitos.info('Ano')
```

```python
df_obitos
```

```python
df_obitos.describe()
```

```python
df_obitos.to_csv('dados_obitos.csv', index=False)
```

```python
df_obitos
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_obitos, x='Ano', y='Óbitos', hue='Causas', palette='Set1')
plt.title('Evolução dos Casos de Óbitos ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Casos')
plt.legend(title='Doença', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
